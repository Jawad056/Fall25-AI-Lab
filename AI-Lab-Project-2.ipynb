```python
class Movie:
    def __init__(self, name, budget):
        self.name = name
        self.budget = budget

class MovieAnalyzer:
    def __init__(self, movies):
        self.movies = movies

    def average_budget(self):
        total = sum(movie.budget for movie in self.movies)
        return total / len(self.movies)

    def above_average_movies(self):
        avg = self.average_budget()
        above_avg = []
        for movie in self.movies:
            if movie.budget > avg:
                difference = movie.budget - avg
                above_avg.append((movie.name, difference))
        return above_avg

    def count_above_average(self):
        return len(self.above_average_movies())
    
movies = [
    Movie("Eternal Sunshine of the Spotless Mind", 20000000),
    Movie("Memento", 9000000),
    Movie("Requiem for a Dream", 4500000),
    Movie("Pirates of the Caribbean: On Stranger Tides", 379000000),
    Movie("Avengers: Age of Ultron", 365000000),
    Movie("Avengers: Endgame", 356000000),
    Movie("Incredibles 2", 200000000)
]

analyzer = MovieAnalyzer(movies)

avg = analyzer.average_budget()
print("Average Budget:", avg)

for name, diff in analyzer.above_average_movies():
    print(name, "is above average by", diff)

print("\nNumber of movies above average: ", analyzer.count_above_average)
```
